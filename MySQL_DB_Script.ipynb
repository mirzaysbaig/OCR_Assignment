```python
import mysql.connector
print("Connection Sucessful")
import json
```

```python
#Connecting to Mysql 
conn = mysql.connector.connect(
    host="localhost",
    user="root", 
    password="Binyamin@321",  
    database="form_data"
)
cursor = conn.cursor()

# Reading JSON data
with open("form_data.json", "r") as file:
    form_data = json.load(file)

# Inserting the patient info 
patient_name = form_data.get("patient_name", "Unknown")
dob = form_data.get("dob", None)

cursor.execute("INSERT INTO patients (name, dob) VALUES (%s, %s)", (patient_name, dob))
patient_id = cursor.lastrowid  

# inserting json data into form data table
json_data = json.dumps(form_data)
cursor.execute("INSERT INTO forms_data (patient_id, form_json) VALUES (%s, %s)", (patient_id, json_data))

#  Commiting now and closing the  connection
conn.commit()
cursor.close()
conn.close()

print("Data inserted successfully!")
```
